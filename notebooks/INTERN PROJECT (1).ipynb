INVALID OR OUT OF RANGE VALUES

```python
df["Gold_Price_USD_per_oz"].describe()
```

```python
df["Gold_Open_USD"].describe()
```

```python
df["Gold_High_USD"].describe()
```

```python
Q1 = df["Gold_Price_USD_per_oz"].quantile(0.25)
Q3 = df["Gold_Price_USD_per_oz"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(lower)
print(upper)
```

```python
outliers = df[
    (df["Gold_Price_USD_per_oz"] < lower) |
    (df["Gold_Price_USD_per_oz"] > upper)
]

outliers
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df["Gold_Price_USD_per_oz"])
plt.show()
```

Here i checked the numerical columns(Gold_prize, gold_open and gold_High) for invalid and out-of range values. Used describe() to see min, max , mean etc.They are no negative and invalid values. All are Positive. I used IQR method to identify outliers. There are Some Outliers. As this is financial data gold prices changes over time, sometimes there will be gaps in prices. So we cannot remove or handle ouliers like we do with other data.
